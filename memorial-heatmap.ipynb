### Memorials & Synagogues Scatter Map

```python
import collections
collections.Iterable = collections.abc.Iterable
```

```python
memorial_df = coordinatesV2[coordinatesV2['type'] == 'Memorial']
synagogue_df = coordinatesV2[coordinatesV2['type'] == 'Synagogue']
```

```python
memorial_df = memorial_df[['latitude','longitude']]
memorial_df['latitude'] = pd.to_numeric(memorial_df['latitude'])
memorial_df['longitude'] = pd.to_numeric(memorial_df['longitude'])
```

```python
synagogue_df = synagogue_df[['latitude','longitude']]
synagogue_df['latitude'] = pd.to_numeric(synagogue_df['latitude'])
synagogue_df['longitude'] = pd.to_numeric(synagogue_df['longitude'])
```

```python
memorial_layer = gmaps.symbol_layer(
    memorial_df, fill_color='rgba(0, 150, 0, 0.4)',
    stroke_color='rgba(0, 150, 0, 0.4)', scale=4
)
```

```python
synagogue_layer = gmaps.symbol_layer(
    synagogue_df, fill_color='rgba(200, 0, 0, 0.4)',
    stroke_color='rgba(200, 0, 0, 0.4)', scale=4
)
```

```python
fig2 = gmaps.figure()
fig2.add_layer(memorial_layer)
fig2.add_layer(synagogue_layer)
fig2
```
